```python
import torch
import torch.nn as nn
from torch.autograd import Variable

from fast_jtnn import *

import numpy as np

import rdkit
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from PIL import Image
```

```python
ls
```

```python
vocab = [x.strip("\r\n ") for x in open("./data/zinc/vocab.txt")]
vocab = Vocab(vocab)

hidden_size = 450
latent_size = 56
depthT = 20
depthG = 3
```

```python
model = JTNNVAE(vocab, hidden_size, latent_size, depthT, depthG)
model.load_state_dict(torch.load("./fast_molvae/new_model/model_25000/model.iter-25000"))
model = model.cuda()
```

```python
np.random.seed(0)
x = np.random.randn(latent_size)
x /= np.linalg.norm(x)

y = np.random.randn(latent_size)
y -= y.dot(x) * x
y /= np.linalg.norm(y)

#z0 = "CN1C(C2=CC(NC3C[C@H](C)C[C@@H](C)C3)=CN=C2)=NN=C1"
z0 = "COC1=CC(OC)=CC([C@@H]2C[NH+](CCC(F)(F)F)CC2)=C1"
z0 = model.encode_latent_from_smiles([z0]).squeeze()

z0 = z0.data.cpu().numpy()
```

```python
z0.shape
```

```python
delta = 1
nei_mols = []
for dx in xrange(-3,4):
    for dy in xrange(-3,4):
        z = z0 + x * delta * dx + y * delta * dy
        tree_z, mol_z = torch.Tensor(z).unsqueeze(0).chunk(2, dim=1)
        tree_z, mol_z = create_var(tree_z), create_var(mol_z)
        nei_mols.append( model.decode(tree_z, mol_z, prob_decode=False) )

nei_mols = [Chem.MolFromSmiles(s) for s in nei_mols]
img = Draw.MolsToGridImage(nei_mols, molsPerRow=7)
```
